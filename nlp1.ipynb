# **Creación de Diccionario**

### Conteo de palabras

```python
import nltk
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from collections import Counter
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

Lectura de archivo de texto.

```python
with open('book_sample.txt', 'r') as file:
    text = file.read()
```

Para hacer el conteo de palabras se hace la tokenización de palabras, es decir, se divide el texto en palabras individuales. Para dividir el texto se usa la biblioteca NLTK (Natural Language Toolkit) y cada palabra se convierte en un elemento de la lista.

```python
words = nltk.word_tokenize(text)
words = [word.lower() for word in words if word.isalpha()]
```

Se eliminan las palabras comunes que la mayoría de las veces no aportan un significado importante y posteriormente se crea una nueva lista con las palabras del texto que no esten dentro de la lista de palabras comunes.

```python
ignored_words = set(stopwords.words('spanish'))
words = [word for word in words if word not in ignored_words]
```

Con la clase 'Counter' se cuenta la frecuencia en la que aparece una palabra en la lista y se asigna a la variable 'counts' la palabra junto con su frecuencia.

```python
counts = Counter(words)
```

Después se descompone el objeto Counter en dos listas separadas, una con las palabras y una con las frecuencias correspondientes.

```python
words, frequencies = zip(*counts.items())
```

### Nube de palabras

Se crea la nube de palabras con ayuda de la biblioteca 'wordcloud'.

```python
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(width=700, height=300, background_color='white').generate_from_frequencies(counts)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
